# Joint Distribution and Covariance

## Question 1

A store orders 10 units of a certain item. For each unit sold, it makes $10. For each item that remains unsold, it loses $5. Let X be the number of items sold in a week. If the store knows that X is a discrete uniform random variable, with X taking values {0, 1, ..., 10}, find the expected profit.

```r
X = 0:10

# Probability mass function of a discrete uniform rv.
# In a discrete uniform rv all outcomes are equally likely
pmf = 1/length(X)

profit_func = function(x) {
    # Each sold generates $10
    gross = 10 * x
    # each unsold loses $5
    loss = 5 * (10 - x)
    # profit
    gross - loss
}

# get profit for each value of X
profit = profit_func(X)

# calculate E(X) 
expected_profit = sum(profit * pmf)
print(expected_profit)
```

## Question 2

Suppose that the service time at a call center is exponentially distributed with the average call lasting 10 minutes. If 3 workers are each on a
call, what is the probability that they have each completed their service in 12 minutes? Assume that the length of each call is independent of every
other call. Round answer to four decimal places.


```r
mean_call = 10
time_limit = 12
num_workers = 3

# mean = 1/rate, solve for rate
rate = 1/10

# find probability for one worker
prob_one = pexp(q = time_limit, rate = rate)

prob_all = prob_one ^ num_workers
round(prob_all, 4)
```

## Question 3

Suppose that X and Y are random variables with the joint probability mass function below:

|       | Y = 1 | Y = 4 | Y = 16 |
|-------|-------|-------|--------|
| X = 1 | 0.20  | 0.25  | 0.05   |
| X = 5 | 0.10  | 0.15  | 0.25   |

### A. Find P(X = 1).

```r
# create matrix
joint_pmf = matrix(c(.2, .25, .05,
                    .1, .15, .25),
                    nrow = 2, byrow = TRUE)
rownames(joint_pmf) = c("X=1", "X=5")
colnames(joint_pmf) = c("Y=1", "Y=4", "Y=16")

# sum across the X=1 row
prob_x1 = sum(joint_pmf[1,])

print(prob_x1)
```

### B. Find P(X>Y)

```r
x_values = matrix(c(1,1,1,
            5,5,5),
            nrow = 2, byrow = TRUE)

y_values = matrix(c(1,4,16,
                    1,4,16),
                    nrow = 2, byrow = TRUE)

# find where X > Y
condition = x_values > y_values

prob_x_greater_y = sum(joint_pmf[condition])
print(prob_x_greater_y)
```

### C. Determine Cov(X,Y)

```r
# E[X] = Σ x · P(X = x) = 1 · P(X = 1) + 5 · P(X = 5)
prob_x = rowSums(joint_pmf)
x_val = c(1,5)
E_X = sum(x_val * prob_x)

# E[Y] = Σ y · P(Y = y) = 1 · P(Y = 1) + 4 · P(Y = 4) + 16 · P(Y = 16)
prob_y = colSums(joint_pmf)
y_val = c(1,4,16)
E_Y = sum(y_val * prob_y)

# E[XY]
E_XY = sum(x_values * y_values * joint_pmf)

# Cov(X, Y) = E[XY] - E[X]·E[Y]
cov_xy = E_XY - E_X * E_Y
print(cov_xy)
```

### D. Find ρ, the correlation coefficient of X and Y. Round answer to four decimal places.

```r
# cor(X, Y) = cov(X, Y) / (stdev(x), stdev(y))

# Calculate Var(X) and stdev(x)
#   Var(x) = E(X^2) - E(X)^2
E_X2 = sum(x_val^2 * prob_x)
var_x = E_X2 - E_X^2
stdev_x = sqrt(var_x)

E_Y2 = sum(y_val^2 * prob_y)
var_y = E_Y2 - E_Y^2
stdev_y = sqrt(var_y)

rho = cov_xy / (stdev_x * stdev_y)
round(rho, 4)
```

### E.  Are X and Y independent? (Answer "Yes", "No", or "Can't determine")

Two random variables X and Y are independent if and only if:
P(X = x, Y = y) = P(X = x) · P(Y = y) for all values of x and y
In other words, the joint probability must equal the product of the marginal probabilities for every cell in the table.

if rho != 0, the are not independent. in our previous part, we determined that rho is not 0. Therefore, the answer is "No"
