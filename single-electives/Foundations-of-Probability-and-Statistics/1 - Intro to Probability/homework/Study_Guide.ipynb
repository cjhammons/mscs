# Probability and Statistics Study Guide

## 1. Axioms of Probability

### Basic Axioms\n
1.  **Non-negativity:** For any event A, P(A) ≥ 0.\n
2.  **Additivity:** If A and B are disjoint events, then P(A ∪ B) = P(A) + P(B).\n
3.  **Normalization:** The probability of the entire sample space S is 1, i.e., P(S) = 1.

### Complements, Unions, and Intersections

```r
# P(A^c) = 1 - P(A)\n
# P(A U B) = P(A) + P(B) - P(A n B)\n
# P(A) = P(A n B) + P(A n B^c)
```

### Principle of Inclusion-Exclusion

```r
# For three events A, B, and C:\n
# P(A U B U C) = P(A) + P(B) + P(C) - P(A n B) - P(A n C) - P(B n C) + P(A n B n C)
```

### Combinatorics

```r
# Permutations: nPr = n! / (n-k)!
# Combinations: nCr = n! / (k!(n-k)!)
factorial(5) # 5!\n
choose(5, 2) # 5C2
```

## 2. Conditional Probability

```r
# P(A|B) = P(A n B) / P(B)
```

### Law of Total Probability

```r
# P(A) = P(A|B)P(B) + P(A|B^c)P(B^c)
```

### Bayes' Theorem

```r
# P(A|B) = [P(B|A)P(A)] / P(B)
```

### Independence of Events

```r
# P(A n B) = P(A)P(B)\n
# P(A|B) = P(A)
```

## 3. Discrete Random Variables

### PMF, E(X), Var(X)

```r
# E(X) = sum(x * P(X=x))\n
# Var(X) = E(X^2) - [E(X)]^2
```

### Binomial Distribution

```r
# probability mass function
dbinom(x, size, prob) # P(X=x)\n
# cumulative distribution function
pbinom(q, size, prob) # P(X<=q)\n
# E(X) = np\n
# Var(X) = np(1-p)
```

### Geometric Distribution

```r
dgeom(x, prob) # P(X=x) (x = # of failures)\n
# E(X) = 1/p (number of trials)\n
# Var(X) = (1-p)/p^2
```

### Negative Binomial Distribution

```r
dnbinom(x, size, prob) # P(X=x) (x = # of failures)\n
# E(X) = r/p (number of trials)\n
# Var(X) = r(1-p)/p^2
```

### Poisson Distribution

```r
dpois(x, lambda) # P(X=x)\n
ppois(q, lambda) # P(X<=q)\n
# E(X) = lambda\n
# Var(X) = lambda
```

## 4. Continuous Random Variables

### PDF, CDF, E(X), Var(X)

```r
# E(X) = integral(x * f(x)) dx\n
# Var(X) = E(X^2) - [E(X)]^2
```

### Uniform Distribution

```r
# E(X) = (a+b)/2\n
# Var(X) = (b-a)^2/12
```

### Exponential Distribution

```r
pexp(q, rate)\n
# E(X) = 1/lambda\n
# Var(X) = 1/lambda^2
```

### Normal Distribution

```r
pnorm(q, mean, sd)\n
qnorm(p, mean, sd)
```

## 5. Joint Distributions

```r
# Cov(X,Y) = E(XY) - E(X)E(Y)\n
# Cor(X,Y) = Cov(X,Y) / (sd(X)sd(Y))
```

### Linear Combinations

```r
# E(aX + bY) = aE(X) + bE(Y)\n
# Var(aX + bY) = a^2*Var(X) + b^2*Var(Y) + 2ab*Cov(X,Y)
```
