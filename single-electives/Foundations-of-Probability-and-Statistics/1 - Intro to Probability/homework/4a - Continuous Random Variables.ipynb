## Question 1

A skilled worker requires at least 10 minutes, and no more than 20 minutes, to complete a certain task. The completion time $X$ is a continuous random variable with density function $f(x) = c/x^2$ for $10 \le x \le 20$ and $f(x)=0$ for all other values of $x$.

What is the value of $c$?

```r
# For f(x) to be a valid probability density function, its integral over its support must be equal to 1.
# So, we need to solve the integral of c/x^2 from 10 to 20 for c.
# Integral(c/x^2) dx = -c/x
# Evaluating from 10 to 20: (-c/20) - (-c/10) = -c/20 + 2c/20 = c/20
# We set this equal to 1: c/20 = 1 => c = 20

# We can verify this using R's integrate function.
pdf = function(x) { 1/x^2 }
integral_val = integrate(pdf, lower = 10, upper = 20)$value
c = 1 / integral_val
c
```

## Question 2

Find the probability that the worker completes the task in 15 minutes or less. Round your answer to three decimal places.

```r
# We need to calculate P(X <= 15). Since the support starts at 10, this is the integral of f(x) from 10 to 15.
# f(x) = 20/x^2
# Integral(20/x^2) dx = -20/x
# Evaluating from 10 to 15: (-20/15) - (-20/10) = -4/3 + 2 = 2/3

c = 20
pdf_c = function(x) { c / x^2 }
prob = integrate(pdf_c, lower = 10, upper = 15)$value
round(prob, 3)
```

## Question 3

Find the expected time for the worker to complete the task. Round your answer to three decimal places.

```r
# The expected value E(X) is the integral of x * f(x) over the support.
# E(X) = Integral(x * (20/x^2)) dx = Integral(20/x) dx from 10 to 20.
# The integral of 20/x is 20 * ln(x).
# Evaluating from 10 to 20: 20*ln(20) - 20*ln(10) = 20 * (ln(20) - ln(10)) = 20 * ln(20/10) = 20 * ln(2).

c = 20
integrand = function(x) { x * c / x^2 }
expected_time = integrate(integrand, lower = 10, upper = 20)$value
round(expected_time, 3)
```

## Question 4

The number of eggs laid on a tree leaf by an insect of a certain type is a Poisson random variable, $X$, with parameter $c$, where $E(X)=c$. However, such a random variable can only be observed if it is positive, since if it is 0, then we cannot know that such an insect was on the leaf. If we let $Y$ denote the observed number of eggs, then $P(Y=i)=P(X=i|X>0)$, where $X$ is Poisson with parameter $c$.

Find $E(Y)$.

**Answer:** $c / (1 - e^{-c})$

**Explanation:**
By definition, $E[Y] = \sum_{i=1}^{\infty} i \cdot P(Y=i)$.
We are given $P(Y=i) = P(X=i | X>0) = \frac{P(X=i \text{ and } X>0)}{P(X>0)}$.
Since $i$ starts from 1, the condition $X>0$ is automatically satisfied, so $P(X=i \text{ and } X>0) = P(X=i)$.
For a Poisson distribution, $P(X=0) = \frac{e^{-c}c^0}{0!} = e^{-c}$.
Therefore, $P(X>0) = 1 - P(X=0) = 1 - e^{-c}$.
Substituting back, $P(Y=i) = \frac{P(X=i)}{1 - e^{-c}}$.
Now, $E[Y] = \sum_{i=1}^{\infty} i \cdot \frac{P(X=i)}{1 - e^{-c}} = \frac{1}{1 - e^{-c}} \sum_{i=1}^{\infty} i \cdot P(X=i)$.
The expected value of the original Poisson variable is $E[X] = c = \sum_{i=0}^{\infty} i \cdot P(X=i)$.
$E[X] = 0 \cdot P(X=0) + \sum_{i=1}^{\infty} i \cdot P(X=i) = \sum_{i=1}^{\infty} i \cdot P(X=i)$.
So, the sum we need is exactly $E[X] = c$.
Finally, $E[Y] = \frac{c}{1 - e^{-c}}$.

## Question 5

Suppose $X$ is a random variable $X \sim N(12, 4)$.

Find the probability that $X$ is within 1.5 standard deviations of the mean. Round your answer to four decimal places.

```r
# The distribution is Normal with mean (mu) = 12 and variance (sigma^2) = 4.
# The standard deviation (sigma) is the square root of the variance.
mu = 12
variance = 4
sigma = sqrt(variance)

# We want to find the probability that X is within 1.5 standard deviations of the mean.
# This is the interval [mu - 1.5*sigma, mu + 1.5*sigma].
lower_bound = mu - 1.5 * sigma
upper_bound = mu + 1.5 * sigma

# We need to find P(lower_bound <= X <= upper_bound).
# This is P(X <= upper_bound) - P(X < lower_bound).
prob = pnorm(upper_bound, mean = mu, sd = sigma) - pnorm(lower_bound, mean = mu, sd = sigma)
round(prob, 4)

# Alternatively, for any normal distribution, the probability of being within z standard deviations
# is the same as a standard normal variable Z being between -z and z.
round(pnorm(1.5) - pnorm(-1.5), 4)
```

## Question 6

Find $k$ such that $P(X>k)=0.10$. Round your answer to two decimal places.

```r
# We are given P(X > k) = 0.10. This is the upper tail probability.
# The qnorm function in R works with the lower tail probability, P(X <= k).
# P(X <= k) = 1 - P(X > k) = 1 - 0.10 = 0.90.

mu = 12
sigma = 2
p_lower_tail = 0.90

# We use the qnorm function to find the value k for which the cumulative probability is 0.90.
k = qnorm(p_lower_tail, mean = mu, sd = sigma)
round(k, 2)

# Alternatively, we can tell qnorm to use the upper tail directly.
k_upper = qnorm(0.10, mean = mu, sd = sigma, lower.tail = FALSE)
round(k_upper, 2)
```
