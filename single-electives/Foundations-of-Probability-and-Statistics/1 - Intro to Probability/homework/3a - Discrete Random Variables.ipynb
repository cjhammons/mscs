## Question 1

A critical system has 5 different components, each of which works with probability 0.90. Assume each component works independently of the others. Let $X$ be the number of components that are working.

Find $P(X=2)$. Round your answer to four decimal places.

```r
# This is a binomial distribution problem because we have a fixed number of independent trials (5 components),
# each with two possible outcomes (working or not working), and a constant probability of success (0.90).

# n = 5 (number of trials/components)
# p = 0.90 (probability of success/working)
# k = 2 (number of successes we are interested in)

# We use the dbinom function in R to find the probability of exactly k successes.
prob = dbinom(x = 2, size = 5, prob = 0.90)
round(prob, 4)
```

## Question 2

Find $E(X)$.

```r
# The expected value (mean) of a binomial distribution is given by the formula E(X) = n * p.
n = 5
p = 0.90
expected_value = n * p
expected_value
```

## Question 3

Find $V(X)$. Round your answer to two decimal places.

```r
# The variance of a binomial distribution is given by the formula V(X) = n * p * (1 - p).
n = 5
p = 0.90
variance = n * p * (1 - p)
round(variance, 2)
```

## Question 4

A certain type of item produced by a factory has a 6% chance of being defective. Draw a random sample until you get the first defective. Let $X$ be the number of items that are drawn.

Find $P(X=2)$. Round your answer to four decimal places.

```r
# This is a geometric distribution because we are looking for the number of trials required to get the first success.
# The probability of success (finding a defective item) is p = 0.06.
# P(X=k) = (1-p)^(k-1) * p

# For P(X=2), it means the first item was not defective (failure) and the second was (success).
p = 0.06
prob_x_2 = (1 - p)^(2 - 1) * p
round(prob_x_2, 4)

# Alternatively, using R's dgeom function. Note that dgeom's x is the number of failures before the first success.
# So, for X=2, there is x=1 failure.
round(dgeom(x = 1, prob = 0.06), 4)
```

## Question 5

Find $E(X)$. Round answer to two decimal places.

```r
# The expected value of a geometric distribution is E(X) = 1 / p.
p = 0.06
expected_value_geom = 1 / p
round(expected_value_geom, 2)
```

## Question 6

Find $V(X)$. Round answer to two decimal places.

```r
# The variance of a geometric distribution is V(X) = (1 - p) / p^2.
p = 0.06
variance_geom = (1 - p) / p^2
round(variance_geom, 2)
```

## Question 7

A certain system can experience three different types of defects. Let $A_i$ for $i = 1, 2, 3$ be the event that the system has a defect of type $i$. Suppose that $P(A_1)=0.17$, $P(A_2)=0.07$, $P(A_3)=0.13$, $P(A_1 \cup A_2)=0.18$, $P(A_2 \cup A_3)=0.18$, $P(A_1 \cup A_3)=0.19$, and $P(A_1 \cap A_2 \cap A_3)=0.01$. Let the random variable $X$ be the number of defects that are present.

Calculate $P(X=0)$.

```r
# First, define all the given probabilities.
p_a1 = 0.17
p_a2 = 0.07
p_a3 = 0.13
p_a1_cup_a2 = 0.18
p_a2_cup_a3 = 0.18
p_a1_cup_a3 = 0.19
p_a1_cap_a2_cap_a3 = 0.01

# P(X=0) is the probability of no defects, which is the complement of at least one defect.
# P(X=0) = 1 - P(A1 U A2 U A3)

# We need to find the intersections of pairs first using P(A U B) = P(A) + P(B) - P(A n B)
p_a1_cap_a2 = p_a1 + p_a2 - p_a1_cup_a2
p_a2_cap_a3 = p_a2 + p_a3 - p_a2_cup_a3
p_a1_cap_a3 = p_a1 + p_a3 - p_a1_cup_a3

# Now use the Principle of Inclusion-Exclusion for three events.
p_a1_cup_a2_cup_a3 = (p_a1 + p_a2 + p_a3) - (p_a1_cap_a2 + p_a2_cap_a3 + p_a1_cap_a3) + p_a1_cap_a2_cap_a3

p_x_0 = 1 - p_a1_cup_a2_cup_a3
p_x_0
```

## Question 8

Calculate $P(X=1)$.

```r
# P(X=1) is the probability of having exactly one defect.
# P(X=1) = P(A1 only) + P(A2 only) + P(A3 only)
# P(A1 only) = P(A1) - P(A1nA2) - P(A1nA3) + P(A1nA2nA3)
p_a1_only = p_a1 - p_a1_cap_a2 - p_a1_cap_a3 + p_a1_cap_a2_cap_a3
p_a2_only = p_a2 - p_a1_cap_a2 - p_a2_cap_a3 + p_a1_cap_a2_cap_a3
p_a3_only = p_a3 - p_a1_cap_a3 - p_a2_cap_a3 + p_a1_cap_a2_cap_a3

p_x_1 = p_a1_only + p_a2_only + p_a3_only
p_x_1

# An alternative way is to use the fact that probabilities must sum to 1, but we need P(X=2) and P(X=3) first.
```

## Question 9

Calculate $P(X=2)$.

```r
# P(X=2) is the probability of having exactly two defects.
# P(X=2) = P(A1 and A2 only) + P(A1 and A3 only) + P(A2 and A3 only)
# P(A1 and A2 only) = P(A1 n A2) - P(A1 n A2 n A3)
p_a1_a2_only = p_a1_cap_a2 - p_a1_cap_a2_cap_a3
p_a1_a3_only = p_a1_cap_a3 - p_a1_cap_a2_cap_a3
p_a2_a3_only = p_a2_cap_a3 - p_a1_cap_a2_cap_a3

p_x_2 = p_a1_a2_only + p_a1_a3_only + p_a2_a3_only
p_x_2
```

## Question 10

Calculate $P(X=3)$.

```r
# P(X=3) is the probability of having all three defects.
# This is simply the probability of the intersection of all three events.
p_x_3 = p_a1_cap_a2_cap_a3
p_x_3
```

## Question 11

Find $E(X)$.

```r
# The expected value E(X) is the sum of each outcome multiplied by its probability.
# E(X) = sum(x * P(X=x))
e_x = (0 * p_x_0) + (1 * p_x_1) + (2 * p_x_2) + (3 * p_x_3)
e_x

# A simpler method is to use the linearity of expectation.
# Let I_i be an indicator variable for event A_i.
# X = I_1 + I_2 + I_3
# E(X) = E(I_1) + E(I_2) + E(I_3) = P(A1) + P(A2) + P(A3)
e_x_simple = p_a1 + p_a2 + p_a3
e_x_simple
```

## Question 12

Find $V(X)$. Give four decimal places for your answer.

```r
# The variance V(X) is calculated as E(X^2) - [E(X)]^2.
# First, calculate E(X^2).
e_x_squared = (0^2 * p_x_0) + (1^2 * p_x_1) + (2^2 * p_x_2) + (3^2 * p_x_3)

# Now, calculate the variance.
v_x = e_x_squared - e_x^2
round(v_x, 4)
```

## Question 13

Find $\sigma$, the standard deviation of $X$. Round your answer to three decimal places.

```r
# The standard deviation is the square root of the variance.
std_dev_x = sqrt(v_x)
round(std_dev_x, 3)
```
