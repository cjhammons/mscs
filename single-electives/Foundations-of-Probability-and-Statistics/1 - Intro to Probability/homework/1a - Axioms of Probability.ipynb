## Question 1

For each of the following scenarios, find the requested probability. Assume the sets A, B, and C are events from the same sample space S. (Hint: Venn diagrams may help with the visualization, although they are not required to answer the questions.)

If $P(A)=.4$, $P(B^c)=.7$, and $P(A \cap B^c)=.2$, find $P(A \cap B)$

```r
p_a = 0.4
p_bc = 0.7
p_a_intersect_bc = 0.2

# The Axiom of probability states $P(A) = P(A \cap B) + P(A \cap B^c)$
# We can rearrange this to solve for $P(A \cap B)$
p_a_intersect_b = p_a - p_a_intersect_bc
p_a_intersect_b
```

## Question 2

For each of the following scenarios, find the requested probability. Assume the sets A, B, and C are events from the same sample space S. (Hint: Venn diagrams may help with the visualization, although they are not required to answer the questions.)

If $P(A)=0.9$ and $P(B)=0.9$, what is the lower bound for $P(A \cup B)$.

```r
p_a = 0.9
p_b = 0.9

# We know that $P(A \cup B)$ must be at least as large as the largest of P(A) and P(B).
# Since $P(A)$ and $P(B)$ are both 0.9, the lower bound for $P(A \cup B)$ is 0.9.
lower_bound = max(p_a, p_b)
lower_bound
```

## Question 3

Three popular options on a certain type of car are A leather seats, B a sunroof, and C heated seats. In the past, $P(A)=0.55$ (i.e. 55% of the customers have requested option A), $P(B)=0.45$, $P(C)=0.4$. Furthermore, $P(A \cap B)=0.25$, $P(A \cap C)=0.2$, $P(B \cap C)=0.15$ and $P(A \cap B \cap C)=0.1$.

Find the probability that a customer will ask for at least one of the three options.

```r
p_a = 0.55
p_b = 0.45
p_c = 0.4
p_a_int_b = 0.25
p_a_int_c = 0.2
p_b_int_c = 0.15
p_a_int_b_int_c = 0.1

# Using the Principle of Inclusion-Exclusion for three sets:
# $P(A \cup B \cup C) = P(A) + P(B) + P(C) - P(A \cap B) - P(A \cap C) - P(B \cap C) + P(A \cap B \cap C)$
p_at_least_one = p_a + p_b + p_c - p_a_int_b - p_a_int_c - p_b_int_c + p_a_int_b_int_c
p_at_least_one
```

## Question 4

Find the probability that a customer will not ask for any of these three options.

```r
# The probability of a customer not asking for any option is the complement of asking for at least one.
# $P(\text{none}) = 1 - P(\text{at least one})$
p_none = 1 - p_at_least_one
p_none
```

## Question 5

Find the probability that a customer will ask for heated leather seats but not a sunroof.

```r
# This corresponds to $P(A \cap C \cap B^c)$.
# We know that $P(A \cap C) = P(A \cap C \cap B) + P(A \cap C \cap B^c)$.
# Therefore, $P(A \cap C \cap B^c) = P(A \cap C) - P(A \cap B \cap C)$.
p_a_c_not_b = p_a_int_c - p_a_int_b_int_c
p_a_c_not_b
```

## Question 6

Find the probability that a customer will ask for at most two of the options.

```r
# Asking for 'at most two' is the complement of asking for all three options.
# $P(\text{at most two}) = 1 - P(A \cap B \cap C)$
p_at_most_two = 1 - p_a_int_b_int_c
p_at_most_two
```

## Question 7

Find the probability that a customer will ask for exactly two of the options.

```r
# $P(\text{exactly two}) = P(A \cap B \cap C^c) + P(A \cap C \cap B^c) + P(B \cap C \cap A^c)$
p_a_b_not_c = p_a_int_b - p_a_int_b_int_c
p_a_c_not_b = p_a_int_c - p_a_int_b_int_c
p_b_c_not_a = p_b_int_c - p_a_int_b_int_c

p_exactly_two = p_a_b_not_c + p_a_c_not_b + p_b_c_not_a
p_exactly_two
```

## Question 8

A message of length 5 digits is to be sent. Each digit can be a 0, 1, or 2.

What is the cardinality of the sample space?

```r
# There are 3 choices for each of the 5 positions.
n_choices = 3
n_digits = 5
sample_space_size = n_choices ^ n_digits
sample_space_size
```

## Question 9

If every message is equally likely, what is the probability that the message consists of 2 zeros, 2 ones, and 1 two? Round your answer to have four decimal places.

```r
# The number of distinct permutations of the 5 digits (2 zeros, 2 ones, 1 two) can be found using the multinomial coefficient formula:
# $$\frac{n!}{n_1! n_2! ... n_k!}$$
favorable_outcomes = factorial(5) / (factorial(2) * factorial(2) * factorial(1))

# The total number of possible messages is the sample space size from the previous question.
total_outcomes = 3^5

probability = favorable_outcomes / total_outcomes
round(probability, 4)
```

## Question 10

What is the probability that the message contains at least one zero? Round your answer to have three decimal places.

```r
# This is the complement of the event that the message contains no zeros.
# If a message has no zeros, each of the 5 digits can only be a 1 or a 2 (2 choices).
outcomes_no_zeros = 2^5
total_outcomes = 3^5

prob_no_zeros = outcomes_no_zeros / total_outcomes
prob_at_least_one_zero = 1 - prob_no_zeros

round(prob_at_least_one_zero, 3)
```
