## Question 1

Suppose 30 numbers are selected at random from the interval [0,1]. That is, $X_i \sim U[0,1]$ for $i=1,2,...30$. Given $\bar{X}=\frac{1}{30}\sum_{i=1}^{30}X_i$ 

estimate $P(0.5 \le \bar{X} \le 0.6)$. (Round your answer to three decimal places.)

```r
n = 30
a = 0
b = 1

# mean of uniform
mu = (a + b) / 2
# var of uniform
var = (b - a) ^ 2 / 12

# mean of sample's distribution
mu_x_bar = mu

# variance of sample mean's distribution
variance_x_bar = var / n 

sd_x_bar = sqrt(variance_x_bar)

prob = pnorm(0.6, mean = mu_x_bar, sd = sd_x_bar) - pnorm(.5, mean = mu_x_bar, sd = sd_x_bar)
round(prob, 3)
```

## Question 2

A company wants to compare the efficiency of two types of fuel for a car. 5 identical cars will be driven 500 kilometers each, two with the first type of fuel and three with the second type. Let $X_1$ and $X_2$ be the observed fuel efficiency for the first type and $Y_1, Y_2$, and $Y_3$ be the efficiency for the second type. Suppose these variables are independent and $X_i \sim N(20,4)$ for $i=1,2$ and $Y_j \sim N(18,9)$ for $j=1,2,3$. Define a new random variable: $W=\frac{X_1+X_2}{2} - \frac{Y_1+Y_2+Y_3}{3}$.

### A. Find $E(W)$.

```r
e_x1 = e_x2 = 20
e_y1 = e_y2 = e_y3 = 18

e_w = ((e_x1 + e_x2) / 2) - ((e_y1 + e_y2 + e_y3) / 3)
e_w
```

### B. Find $Var(W)$.

```r
var_x1 = var_x2 = 4
var_y1 = var_y2 = var_y3 = 9

# pull the contants out and square
var_w = (1/2)^2 * (var_x1 + var_x2) + (1/3)^2 *(var_y1 + var_y2 + var_y3)
var_w
```
