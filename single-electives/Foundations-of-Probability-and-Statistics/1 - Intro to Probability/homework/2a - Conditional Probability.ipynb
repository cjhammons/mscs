## Question 1

If $A$ and $B$ are two events such that $P(A)=0.7$, $P(B)=0.5$, and $P(B|A)=0.6$, find $P(A \cup B)$. If possible, round your answer to two decimal places.

```r
p_a = 0.7
p_b = 0.5
p_b_given_a = 0.6

# Formula for the union of two events: $P(A \cup B) = P(A) + P(B) - P(A \cap B)$
# First, find $P(A \cap B)$ using the conditional probability formula: $P(B|A) = P(A \cap B) / P(A)$
# Rearranging gives: $P(A \cap B) = P(B|A) * P(A)$
p_a_intersect_b = p_b_given_a * p_a

# Now calculate $P(A \cup B)$
p_a_union_b = p_a + p_b - p_a_intersect_b
round(p_a_union_b, 2)
```

## Question 2

If $A$ is a subset of $B$, can $A$ and $B$ be independent? (Answer Yes, No, or Can't tell.)

**Answer: No**

**Explanation:**
For two events to be independent, $P(A \cap B) = P(A)P(B)$.
If $A$ is a subset of $B$, then the intersection of $A$ and $B$ is $A$ itself, so $P(A \cap B) = P(A)$.
Substituting this into the independence equation gives $P(A) = P(A)P(B)$.
Since the problem states $0 < P(A) < 1$, we can divide both sides by $P(A)$, which results in $1 = P(B)$.
However, the problem also states $0 < P(B) < 1$, which contradicts $P(B)=1$. Therefore, $A$ and $B$ cannot be independent.

## Question 5

If $P(A) = 0.8$, $P(B|A) = 0.5$ and A and B are dependent (not independent), then is $P(B) \le P(A)$? (Answer Yes, No, or Can't tell)

**Answer: Yes**

**Explanation:**
We know $P(A \cup B) = P(A) + P(B) - P(A \cap B)$.
We can find $P(A \cap B) = P(B|A)P(A) = 0.5 \times 0.8 = 0.4$.
Substituting this in, we get $P(A \cup B) = 0.8 + P(B) - 0.4 = 0.4 + P(B)$.
By the axioms of probability, the probability of any event cannot exceed 1, so $P(A \cup B) \le 1$.
This implies $0.4 + P(B) \le 1$, which means $P(B) \le 0.6$.
Since $P(B) \le 0.6$ and $P(A) = 0.8$, it is true that $P(B) \le P(A)$.

## Question 6

An unfair coin is flipped four times. The probability of a heads is 0.6 and the probability of a tails is 0.4. Every flip is independent of every other flip.

Find the probability of getting exactly 2 heads and 2 tails. Round your answer to four decimal places.

```r
# This is a binomial probability problem.
# n = 4 (number of trials)
# k = 2 (number of successes, i.e., heads)
# p = 0.6 (probability of success)

# The number of ways to get 2 heads in 4 flips is given by the binomial coefficient "4 choose 2".
num_ways = choose(4, 2)

# The probability of any single sequence of 2 heads and 2 tails is p^k * (1-p)^(n-k).
prob_sequence = 0.6^2 * 0.4^2

# The total probability is the number of ways multiplied by the probability of one sequence.
total_prob = num_ways * prob_sequence
round(total_prob, 4)

# Alternatively, use the built-in dbinom function in R.
# dbinom(x=number of successes, size=number of trials, prob=probability of success)
round(dbinom(2, 4, 0.6), 4)
```

## Question 7

Find the probability of getting exactly 2 heads and 2 tails given that the first coin flip was a head. Round your answer to three decimal places.

```r
# Let E be the event of getting exactly 2 heads and 2 tails.
# Let F be the event that the first flip is a head.
# We want to find P(E|F).

# Since the first flip is a head, for event E to occur, we need exactly 1 more head and 2 tails in the remaining 3 flips.
# This simplifies the problem to finding the probability of 1 head in 3 flips.

# n = 3 (remaining trials)
# k = 1 (remaining heads needed)
# p = 0.6 (probability of heads)
prob = dbinom(1, 3, 0.6)
round(prob, 3)
```

## Question 8

70% of the light aircraft that disappear while in flight in a certain country are subsequently found. Of the aircraft that are found, 60% have an emergency locator, whereas only 10% of the aircraft not found have such a locator. Suppose a light aircraft has disappeared.

If the aircraft has an emergency locator, what is the probability that it will not be found? Round your answer to three decimal places.

```r
# Let F be the event that the aircraft is found.
# Let L be the event that the aircraft has a locator.

# Given probabilities:
p_f = 0.70
p_l_given_f = 0.60
p_l_given_not_f = 0.10

# We want to find P(F^c | L).
# Using Bayes' Theorem: P(F^c | L) = (P(L | F^c) * P(F^c)) / P(L)

# First, find the complements:
p_not_f = 1 - p_f

# Next, find P(L) using the Law of Total Probability:
# P(L) = P(L|F)P(F) + P(L|F^c)P(F^c)
p_l = p_l_given_f * p_f + p_l_given_not_f * p_not_f

# Now, apply Bayes' Theorem:
p_not_f_given_l = (p_l_given_not_f * p_not_f) / p_l
round(p_not_f_given_l, 3)
```
